```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error
import os

CONFIG = {
    'input_directory': 'Data/PriceData',
    'output_directory': 'Results',
}

def load_data(file_path):
    df = pd.read_parquet(file_path)
    if 'Date' not in df.columns:
        if df.index.name == 'Date':
            df = df.reset_index()
        else:
            df['Date'] = pd.date_range(start='2000-01-01', periods=len(df))
    df['Date'] = pd.to_datetime(df['Date'])
    return df[['Date', 'Open', 'High', 'Low', 'Close', 'Volume']]

def prepare_data(df):
    max_lag = 7  # Maximum lag used in our features

    df['Future_Returns'] = df['Close'].pct_change(7).shift(-7)
    df['Future_Returns'] = df['Future_Returns'].ewm(span=5).mean()
    




    
    for col in ['Open', 'High', 'Low', 'Close', 'Volume']:
        for i in range(1, 8):
            df[f'{col}{i}'] = df[col].shift(i)
        df[f'{col}_Lag1'] = df[col].shift(3)
        df[f'{col}_Lag2'] = df[col].shift(7)
        df[f'{col}_Pct_Change'] = df[col].pct_change()
    
    # Remove NaN values created by shifting and percentage change calculations
    df = df.dropna()
    
    # Remove the first and last few rows to ensure all lagged features are available
    df = df.iloc[max_lag:-max_lag]
    
    # Further ensure no NaN values in the dataset
    df = df.dropna()
    
    return df

formulas = [
    "(-1) * ((data['Open_Lag2'] - (data['High3'] + data['Close'])) / data['Close'])",
    "np.sin(data['Open_Pct_Change'] + data['High6'] + np.sin(data['Open_Pct_Change'] + data['High6']))",
    "(data['Close'] - data['High1']) / data['Volume']",
    "(data['Close'] - data['High1']) / (data['High_Lag2'] * data['High7'])",
    "np.sin(np.sin((data['Close'] - data['High_Lag2']) / data['High']) + data['High6'])",
    "(data['Close'] - data['High4'] + data['High3']) / data['High7']",
    "(data['Close'] - data['High4'] + data['High3']) / data['Volume']",
    "(data['Close'] - data['High2']) / data['High7']",
    # Additional formulas
    "(data['Close'] - data['Low_Lag1']) / data['Low2']",
    "np.log1p(data['Volume'] / (data['High1'] - data['Low1']))",
    "(data['High3'] - data['Low3']) / data['Close_Lag2']",
    "(data['Close'] - data['Low_Lag2']) / data['Volume']",
    "(data['High2'] - data['Low2']) * data['Close_Pct_Change']",
    "np.tan(data['Open_Lag1'] - data['Low3']) / data['Close']",
    "np.sqrt(np.abs(data['High4'] - data['Low4'])) / data['Volume']",
    "np.exp((data['Close'] - data['High3']) / data['Low3'])",
    "(data['Close'] - data['Open_Lag2']) / (data['High1'] - data['Low1'])",
    "np.sin((data['Close'] - data['Low2']) / data['High1'])"
]


def evaluate_formula(formula, data):
    y_true = data['Future_Returns'].values
    y_pred = eval(formula)
    
    if len(y_true) != len(y_pred):
        min_len = min(len(y_true), len(y_pred))
        y_true = y_true[:min_len]
        y_pred = y_pred[:min_len]
    
    mse = mean_squared_error(y_true, y_pred)
    mae = mean_absolute_error(y_true, y_pred)
    correlation = np.corrcoef(y_true, y_pred)[0, 1]
    direction_accuracy = np.mean((np.sign(y_true) == np.sign(y_pred)))
    baseline_accuracy = np.mean(y_true > 0)
    edge = direction_accuracy - baseline_accuracy

    print(f"MSE: {mse:.6f}")
    print(f"MAE: {mae:.6f}")
    print(f"Correlation: {correlation:.6f}")
    print(f"Direction Accuracy: {direction_accuracy:.6f}")
    print(f"Baseline Accuracy: {baseline_accuracy:.6f}")
    print(f"Edge: {edge:.6f}")
    print(f"Edge over baseline: {edge*100:.2f}%")

    plt.figure(figsize=(12, 5))
    plt.scatter(y_true, y_pred, color='grey', alpha=0.5)
    plt.scatter(0, 0, color='red', s=50)  # Red dot at 0,0
    plt.xlabel('Actual Returns')
    plt.ylabel('Predicted Returns')
    plt.xscale('symlog')
    plt.yscale('symlog')
    plt.title('Predicted vs Actual Returns')
    plt.grid(True)
    plt.show()

def main():
    file_paths = [os.path.join(CONFIG['input_directory'], f) for f in os.listdir(CONFIG['input_directory']) if f.endswith('.parquet')]
    all_data = pd.concat([load_data(file) for file in file_paths[:100]], ignore_index=True)  # Limit to 100 files
    prepared_data = prepare_data(all_data)

    print("Columns in prepared data:")
    print(prepared_data.columns)

    print("\nShape of prepared data:", prepared_data.shape)

    print("\nSample of prepared data:")
    print(prepared_data.head())

    for i, formula in enumerate(formulas, 1):
        print(f"\nEvaluating Formula {i}:")
        print(formula)
        try:
            evaluate_formula(formula, prepared_data)
        except Exception as e:
            print(f"Error evaluating formula: {e}")

if __name__ == "__main__":
    main()
```
